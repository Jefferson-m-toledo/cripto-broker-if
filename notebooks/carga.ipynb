```python
def find_time(s):
    result = re.search('T-(.*).json', s)
    return result.group(1)
```

```python
def find_pair(diretorio, time,s):
    """Encontra tempo gr√°fico
    """
    result = re.search(f'{diretorio}/(.*)-{time}.json', s)
    return result.group(1)
```
