```python
def encontra_objeto_banco(index: str):
    # conexão com mongo
    client = MongoClient("mongodb://localhost:27017/")
    # cria database
    db = client["cripto"]
    # cria coleção
    moedas = db["moedas"]
    # ler dados do banco
    data_from_db = moedas.find_one({"index": index})
    # cria dataframe pandas
    df = pd.DataFrame(data_from_db["Candles"])
    return df
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(
        x=df['Data'],
        open=df['Open'],
        high=df['High'],
        low=df['Low'],
        close=df['Close']
    )])

fig.show()
```
